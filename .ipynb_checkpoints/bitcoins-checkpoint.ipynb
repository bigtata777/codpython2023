```python
def procesado(direccion):
    
    df = pd.read_csv(direccion)
    df['Date'] = pd.to_datetime(df['Date']).apply(lambda x: x.strftime('%Y-%m-%d'))
    df=df[['Date','btc_market_price']]
    df.columns=['ds','y']
    return df
    
```

```python
data2 = data.loc[data.ds >='2017-01-01',]
data2.reset_index(drop=True,inplace=True)
```

```python
data2[['ds','y']].set_index('ds').plot()
```

```python
# Calcular la media móvil de 50 días
data2['50_MA'] = data2['y'].rolling(window=50).mean()
data2['20_MA'] = data2['y'].rolling(window=20).mean()
# Gráfica del precio de cierre y la media móvil
plt.figure(figsize=(14,7))
data2[['ds','y']].set_index('ds').plot(label='Precio de cierre', title='bitcoins Precio de Cierre ')
data2['50_MA'].plot(label='Media Móvil 50 días')
data2['20_MA'].plot(label='Media Móvil 20 días')
plt.legend()
plt.show()
```

```python

# Determina las fechas de intersección
data2['Intersect'] = ((data2['20_MA'] > data2['50_MA']) & 
                     (data2['20_MA'].shift(1) <= data2['50_MA'].shift(1))) | \
                    ((data2['20_MA'] < data2['50_MA']) & 
                     (data2['20_MA'].shift(1) >= data2['50_MA'].shift(1)))
data2
```

```python
intersect_dates = data2[data2['Intersect']]['ds']
```

```python


# Graficar
plt.figure(figsize=(12,6))
plt.plot(data2['ds'], data2['y'], label='Bitcoin', color='blue')
plt.plot(data2['ds'], data2['20_MA'], label='Media Móvil 20 días', color='red')
plt.plot(data2['ds'], data2['50_MA'], label='Media Móvil 50 días', color='green')
for date in intersect_dates:
    plt.axvline(x=date, color='gray', linestyle='--')
plt.title('Bitcoins Precio de Cierre')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.show()
```

```python
# Get hold of the Prophet object
m = Prophet(changepoint_range=0.95)
 
# Fit the data. Remember that prophet expect "ds" and "y" as names for the columns.
m.fit(data2)
 
# We must create a data frame holding dates for our forecast. The periods # parameter 
#counts days as long as the frequency is 'D' for the day. Let's # do a 180 day forecast, 
#approximately half a year.
future = m.make_future_dataframe(periods=21, freq='D')
```

```python
# Create the forecast object which will hold all of the resulting data from the forecast.
forecast = m.predict(future)
forecast
# List the predicted values with a lower and upper band.
```

```python
fig, ax = plt.subplots(figsize=(14, 10)) 
plt.plot(forecast['ds'],forecast['yhat'])
ax.set_title("Mean prices bitcoins", fontsize=28, fontweight="bold")
```
